```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def f(x):
    return -12 * x**4 * np.sin(np.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30

def f_prime(x):
    return -48 * x**3 * np.sin(np.cos(x)) + 12 * x**4 * np.cos(np.cos(x)) * np.sin(x) - 54 * x**2 + 10 * x + 10
```

```python
points_list = [x for x in range(-4,5,1)]
```

```python
roots = fsolve(f, points_list)
critical_points = fsolve(f_prime, points_list)
extrema_values = f(critical_points)
print(roots)
print(critical_points)
```

```python
increasing_intervals = []
decreasing_intervals = []

for i in range(len(critical_points) - 1):
    mid_point = (critical_points[i] + critical_points[i + 1]) / 2
    if f_prime(mid_point) > 0:
        increasing_intervals.append((float(critical_points[i]), float(critical_points[i + 1])))
    elif f_prime(mid_point) < 0:
        decreasing_intervals.append((float(critical_points[i]), float(critical_points[i + 1])))

print(increasing_intervals)

print(decreasing_intervals)
```

```python
x_values = np.linspace(-4, 4, 1000)
f_values = f(x_values)
positive_intervals = x_values[f_values > 0]
negative_intervals = x_values[f_values < 0]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x)', color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.scatter(roots, f(roots), color='red', label='Roots', zorder=5)
plt.scatter(critical_points, extrema_values, color='green', label='Extrema', zorder=5)
plt.title("График функции f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
print("Корни функции:", roots)
print("Критические точки:", critical_points)
print("Значения функции в критических точках:", extrema_values)
print("Интервалы возрастания:", increasing_intervals)
print("Интервалы убывания:", decreasing_intervals)
print("Интервалы, где f(x) > 0:", positive_intervals)
print("Интервалы, где f(x) < 0:", negative_intervals)
```
